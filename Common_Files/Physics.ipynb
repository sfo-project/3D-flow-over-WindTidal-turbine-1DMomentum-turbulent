- The thrust part :
The total thrust of a wind turbine is given by :
$$ T=\frac{1}{2}.\rho.V_0^2.C_t.\pi.(R^2-R^2_{hub})~~~(1) $$

As a consequence the thrust on an annular ring of width at a radius r is:

$$ dT=\frac{1}{2}.\rho.V_0^2.C_t.2\pi.r.dr ~~~(2)$$


Because the rings have a finite thickness it possible to write :
$$ \Delta T=\frac{1}{2}.\rho.V_0^2.C_t.2\pi.r.\Delta r~~~(3) $$


Using equation 1, equation 3 can be rewritten as:
$$ \Delta T=T\frac{2r\Delta r}{R^2-R^2_{hub}}~~~(4)$$


Let us calculate the local density of one ring :
$$ \frac{\Delta T}{\Delta V}=\frac{\Delta T}{2\pi r\delta dr}=\frac{1}{2}.\frac{\rho.V^2.C_t}{\delta}=\frac{T}{\delta\pi (R^2-R^2_{hub})}~~~(5)$$


$\delta\pi (R^2-R^2_{hub})$ is the geometric volume of the disk called $V_{geom}$.



The local thrust density thrust is r independent. 

For each cell within a ring:

$$ T_{cell}=\frac{\Delta T}{\Delta V}V_{cell}~~~(6)$$

	
Equation 6 needs to be scaled, because not all cells are fully within each ring. The scaling factor is:

$$\beta=\frac{V_{geom}}{\Sigma_{\text{All cells}} V_{cell}}~~~(7)$$

The thrust per cell is given by:
	
   
$$ T_{cell}=\frac{\Delta T}{\Delta V}V_{cell}.\beta = T\frac{V_{cell}}{\Sigma_\text{All cells}V_{cell}}~~~(8)$$

- Torque part

The definition of the angular momentum L is given by :
 
$$ L=\sum_{i} r_{i}.m_{i}.v_{i}~~~(9) $$


Let us apply (9) for an annular ring of radius r a mass of dm and width $\delta$ :


$$dL= r.dm.v=r.dm.r.\omega~~~(10)$$



 Moreover,

$$dQ=\frac{dL}{dt}=\frac{dm}{dt}.(wr).(r)=\rho.V.2\pi.r.dr(\omega.r)(r)~~~(13)$$


$V=Vo(1-a)$ and  $a'=\frac{\omega}{2.\Omega} $

then,

$$ dQ=4a'(1-a)2\pi.\frac{1}{2}.V_0.\Omega.\rho.r^3.dr~~~(14) $$


By integrating (14) :

$$ Q=4a'(1-a)2\pi.\frac{1}{2}.V_0.\Omega.\rho.\int_{R_{hub}}^{R} r^3.dr~~~(15)$$

The torque is :

$$Q=a'(1-a)2\pi.\frac{1}{2}.V_0.\Omega.\rho.(R^4-R_{hub}^4)~~~(16)$$



Using equation (14), equation (16) can be rewritten :


$$\Delta Q= \frac{4Qr^3\Delta r}{R^4-R_{hub}^4}~~~(17)$$



Torque per unit volume of annular ring is :

$$\frac{\Delta Q}{\Delta V}= \frac{2Qr^2}{\pi\delta(R^4-R_{hub}^4)}~~~(18)$$

Torque for each cell within the annular ring is:

$$Q_{cell}=\frac{\Delta Q}{\Delta V}.V_{cell}=\frac{2Qr^2 V_{cell}}{\pi\delta(R^4-R_{hub}^4)}~~~(19)$$


Just like in the thrust part equation (19) needs to be scaled by $\gamma$ the torque scale factor :


$$\gamma=\frac{V_{geom}. \frac{1}{2}(R^2-R_{hub}^2)}{\sum_{\text{ cells}} r_{cell}^2.V_{cell}}~~~(20)$$



so :


$$ Q_{cell}=\frac{2Qr^2 V_{cell}.\gamma}{\pi\delta(R^4-R_{hub}^4)}~~~(21)$$



$$ V_{geom}=\pi(R^2-R_{hub}^2)\delta~~~(22) $$


So equation (21) can be rewritten :


$$ Q_{cell}=\frac{Q.r^2_{cell}.V_{cell}}{\sum_{\text{ all  cells}} r_{cell}^2.V_{cell}}~~~(23)$$



The tangential force for a given cell whose cell center lies inside the virtual disk is given by:


$$ F_{\text{tangential cell}}=\frac{Q.r_{cell}.V_{cell}}{\sum_{\text{ all  cells}} r_{cell}^2.V_{cell}} ~~~(24)$$


Equations (8) needs the thrust T obtaining from the lookup table in the column Ct using equation (1) and equation (23) needs the torque Q obtaining in the power column by using : $Q=\frac{Power}{\Omega}$ to calculate $T_{\text{cell}}$ and $F_{\text{tangential cell}}$ and imput them in the RANS equations.

#### 1D Momentum (1DM):
The is the simplest method, which requires a lookup table of coefficients of thrust and torque, as a function of flow speed.  As part of the virtual disk model, the 1D Momentum method can be used to model the effects of a wind turbine. The 1D Momentum method is restricted to the Horizontal Axis Wind Turbine (HAWT). The 1D Momentum method is based upon one-dimensional momentum theory with wake rotation. The method imitates the effect of the presence of a wind turbine by adding a source term to the momentum equations. The source term accounts for both axial and tangential effects. The method requires the performance curves of a wind turbine, that is, the thrust coefficient and the power as functions of wind speed. An inflow velocity plane is used to compute the volume-averaged values for the velocity and the density of air. The inflow velocity plane is placed upstream of the wind turbine. This upstream velocity is used to extract the thrust coefficient and the power values from the performance curves.
The method can be used to study wake effects of a single turbine or of a complete wind park. The 1D Momentum method is particularly useful for examining various MHK turbines plant layouts in to order maximize energy production. That is why we have chosen this last method.
