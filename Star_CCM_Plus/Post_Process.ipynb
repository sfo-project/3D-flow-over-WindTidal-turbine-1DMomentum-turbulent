# CFD Results Post-Process and Validation

**The final step for the development of a CFD simulation for a problem of interest is to process and examine the validity of the obtained numerical results (a.k.a post-processing and validation step). This steps requires an in-depth engineering knowledge about the fundamentals of the physical problem of interest. Thus, the value of investing time in understanding the fundamental physics of the flow, once again is highlighted here. This process is done starting from general inspection on CFD simulation results and ends with more detail validation process.**

In this section the results for setup CFD simulation of flow over a **Scaled Model Marine Hydrokinetic Turbine** is presented, discussed and experimentally validated. The first general rule of thumb to check the convergence of a CFD simulation is to visualize the residuals of key flow field variables, such as continuity, momentum, velocity and etc.. In STAR-CCM+ once user starts to run the simulation, the Residual values for the relative variables, based on the choices of models, are visualized in the solver window. Furthermore, users can manage details of the plotted residuals or add other residuals to be visualized via <span style=style="background-color:lightgrey;">Plots/Residuals</span>.

Fig.1 - Residuals of continuity, momentum and turbulent properties for simulation of flow over the scaled model marine hydrokinetic turbine.

Fig. 1 visualizes the residual values of continuity, momentum and turbulent properties versus the number of iterations. The general decreasing trend of the residual values confirms that the numerical solution of the conservation/governing equations are converging to specific final values. As a rule of thumb reaching the continuity residual of 10E-3 is the first necessary but not sufficient general criteria to confirm the convergence of the CFD simulation. In this study case this criteria is reached after about 1500 iterations and after about 5000 iterations the continuity residuals reached value of 5E-4, which is a solid necessary confirmation of CFD simulation convergence. Users can also define criteria to stop the iteration process once the desired criteria is met during the iteration process. These settings are available at via <span style=style="background-color:lightgrey;">Stopping Criteria\ [right click]</span> path. In the current case study two criteria was set to stop the simulations: 1- Reaching via <span style=style="background-color:lightgrey;">Stopping Criteria\Maximum Steps</span> of 11000 OR 2- Reaching the <span style=style="background-color:lightgrey;">Stopping Criteria\Continuity Criterion</span> of 1E-7. The core idea behind choosing the these criteria was to make sure that despite the low Reynolds number value the required number of time steps is used to reach a convergence. For the current case study the first criterion was reach sooner and iterations were stopped after 11000 iterations as shown in Fig. 1.

The second step is to visualize dimensionless form of the important variables within the CFD domain. This visualization step can be problem specific and should be performed over an adequate plane of choice. For most of the problems contour of normalized velocity or pressure, over a horizontal or vertical plane at the center of the CFD domain, is a good start point. The path to define an adequate probe such as planes or lines is <span style=style="background-color:lightgrey;">Derived Parts/ [right click]</span>. With this path a user can define various type of probes to visualize the obtained numerical results over them. For the current problem user can define a plane in the center of the flume via path of <span style=style="background-color:lightgrey;">Derived Parts/ [right click]/Section/Plane...</span>. For this specific domain the origin of the plane was coincide with the main origin at (0,0,0) and the normal was set to be the Z-axis.

User should now define the normalized variable of interest that needs to be visualized. The normalized streamwise velocity by the inlet velocity can be defined using <span style=style="background-color:lightgrey;">Tools/Field Functions/[right click]/New/Scalar</span>. This path will define a variable named <span style=style="background-color:lightgrey;">User Field Function 1</span> in the tree of progress. User should rename this newly defined variable and edit it further in it's window: Function Name (i.e. Velocity) and put the required formula for the variable of interest with correct syntax under Definition (i.e. $${Velocity}[0]/1.13). In this formula the value **1.13** is the imposed free stream inlet velocity as discussed in the Simulation section. This will calculate the normalized streamwise velocity with free stream velocity within entire CFD domain. It order to visualize the defined normalized flow field variable in the previous step users should use <span style=style="background-color:lightgrey;">Scenes/Scalar Scene 1/Displayers/Scalar 1/Scalar Field</span> and choose the variable **velocity**.

<img src="./Images/velocity_cont.png" width="900" align="middle">
Fig.2 - Contours of normalized streamwise velocity with the free stream velocity over a horizontal plane through the rotor zone of the operating turbine.

Fig. 2 shows that as the flow approaches the simulated turbine's rotating blades, modeled via 1D Linear Momentum, it starts to decelerate. The turbine would extract momentum, via imposed sink of momentum on the rotor zone volume, from the flow and produce a turbulent wake and velocity deficit inside it. The velocity deficit is maximum in the near wake region and become less significant as the flow moves further downstream. The wake recovery process is due to the turbulent mixing inside the wake and also leads to momentum exchange with the free stream flow around the wake at the shear layers formed close to the tip of the turbine blade. Furthermore, Fig. 2 confirms that the 1D Linear Momentum model can not capture the detail of the flow right after the blade in near wake region. This is due to averaging the hydrodynamic effects of the blade over the whole area of the disk in 1D Linear Momentum model. Therefore, the turbulent wake has an axisymmetric shape from the beginning after passing the blade.

The visualized velocity contours in Fig. 2 shows the effect of the turbine nacelle and the imposed no-slip boundary condition on it is captured accurately via the provided mesh resolution and used all y+ function model. The imposed no-slip boundary condition lead into significant flow deceleration on the nacelle wall and right after it. This deceleration region is highlighted with a dark blue region around and right after the place of turbine nacelle. 

**After performing the general post-processing steps, the user need to perform more detail oriented post-processing to confirm the validity of the CFD simulation. This stage can be a state-of-the-art. Comparing the numerical results against the theory (discussed earlier) or any other available experimental results is the final stage of post-processing. It is extremely important to calculate the correct corresponding numerical variables and compare them against the experimental or other numerical data.**

In order to validate the results from this simulations the numerically predicted and experimentally measured velocity deficits at the various stations downstream the turbine, highlighted with solid pink lines in Fig. 2, are compared against each other. To perform this comparison the numerical data were extracted from STAR-CCM+ over the defined line probes at each station downstream and then visualized for comparison against experimental data in MATLAB as shown in Fig. 3 below:

<img src="./Images/velocity_deficit_comparison.PNG" width="800" align="middle">
Fig. 3 - Comparison between the numerically predicted (purple dash line) and experimentally measured (solid black line) velocity field 2 diameters upstream and 2, 3, 5 and 7 diameters downstream of the scale model hydrokinetic turbine.

Visualized numerical and experimental results comparison, in the first panel on left of Fig. 3, are in good agreement at the 2D upstream station of the turbine. This agreement is expected since there is no complexity within the flow field before the turbine rotating blades. The velocity profiles at 2D downstream station, in the second panel from left of Fig. 3, show also an overall good agreement between the 1D Linear Momentum model predictions and the experiments. It should be noted that this station is located exactly downstream of the tapered end of the nacelle, on the surface that a high resolution mesh is provided to capture the effect of the nacelle on the shape of the velocity deficit profiles near it. However, there are some minor disagreement between numerical and experimental results due to the averaging the effect of flow at the root of turbine via 1D Linear Momentum model. The most disagreement between the numerical and experimental results at 2D downstream station happens close to the tip of the blade. At the blade tip the velocity magnitude, vector summation of streamwise and angular velocity, is maximum. This high velocity value and rotation of the blade lead in to formation and shedding of tip vortices into the wake of the flow. However, since 1D Linear momentum is simplifying and averaging the turbine rotation effect, hence it is limited to predict the flow accurately in this region. The foot print of this disagreement consistently exist and is visible at the 3D downstream station downstream as well. In the experimental analysis it is observed that the formed and shed tip vortices as a coherent vortex structure in the wake of turbine break down at about 3D downstream and dissipated via turbulent structures within the flow field. Stations 5D and 7D downstream show that the agreement between simulated velocity profiles and the experimental data picture perfect! As the turbine wake becomes more axisymmetric, the 1D Linear Momentum combined with calibrated turbulent closure model **(K-Omega SST and Turbulent length scale of 10 [cm])** matches the experimental results very well. At this stage one can conclude that this CFD numerical simulation is validated to study steady flow over the scaled model hydrokinetic turbine at specific operating conditions. Therefore, these simulations can be used for similar or more complex problems, such as other turbine operating conditions, array of these turbines or examine similar models of these turbines. This can be done via correct implementation in changing the corresponding solver, models and/or boundary conditions.
